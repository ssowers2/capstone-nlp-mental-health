```python
from pathlib import Path

import pandas as pd
```

```python
# Define the location of the raw data folder
RAW_DATA_DIR = Path.cwd().parent / "data" / "raw"

# Load the datasets
neutral_df = pd.read_csv(RAW_DATA_DIR / "Neutral.csv", encoding="latin1")

depression_df = pd.read_csv(RAW_DATA_DIR / "Depression.csv", encoding="latin1")

suicidal_df = pd.read_csv(
    RAW_DATA_DIR / "Suicadal_tendencies_data.csv", encoding="latin1"
)
```

```python
# Explore dataset 1 to see compatibility before merging
neutral_df.head()
```

```python
# Explore dataset 2 to see compatibility before merging
depression_df.head()
```

```python
# Explore dataset 3 to see compatibility before merging
suicidal_df.head()
```

```python
# Validate data, check the dimensions
print("Neutral:", neutral_df.shape)
print("Depression:", depression_df.shape)
print("Suicidal:", suicidal_df.shape)
```

```python
# Check the column names
print(neutral_df.columns)
print(depression_df.columns)
print(suicidal_df.columns)
```

### Observations

All three datasets contained the expected Label and TEXT columns after cleaning.

The Depression dataset originally contained an extra column named `Unnamed: 2`. Inspection showed the column contained only a single non-null value and was unrelated to the analysis, so it was removed before merging the datasets.

```python
# Check for missing values
print("Neutral")
print(neutral_df.isnull().sum())

print("\nDepression")
print(depression_df.isnull().sum())

print("\nSuicidal")
print(suicidal_df.isnull().sum())
```

```python
# Fill missing labels in the Neutral dataset with 0
neutral_df["Label"] = neutral_df["Label"].fillna(0)
```

```python
neutral_df.isnull().sum()
neutral_df["Label"].value_counts()
```

```python
# Confirmed 9 missing text records are still there.
neutral_df.isnull().sum()
```

### Observations

Missing values were found only in the TEXT column.

Rows containing missing text were removed because machine learning models cannot learn from records that do not contain textual information.

No missing values were present in the Label column after validation.

```python
# Merge the datasets
mental_health_df = pd.concat(
    [neutral_df, depression_df, suicidal_df], ignore_index=True
)
```

```python
# Verify merge
print(mental_health_df.shape)
```

### Observations

After validation and cleaning, the three datasets were merged into a single DataFrame.

The merged dataset contains examples from each mental health category while preserving the original class labels.

```python
# Check missing values in merged dataset
mental_health_df.isnull().sum()
```

```python
# Remove rows with missing text
mental_health_df = mental_health_df.dropna(subset=["TEXT"])
```

```python
# Check missing values
mental_health_df.isnull().sum()
```

```python
# Check class distribution
mental_health_df["Label"].value_counts()
```

### Observations

The merged dataset contains three classes:

- Neutral: 6,065 records
- Depression: 4,573 records
- Suicidal Tendencies: 7,061 records

The classes are reasonably balanced, although the Depression class contains fewer observations than the other two categories. This level of imbalance is relatively small and should still allow supervised classification models to learn meaningful patterns.

### Data Quality Summary

The dataset required minimal cleaning prior to analysis.

Data quality issues included:

- One unnecessary column in the Depression dataset.
- Missing text values in each dataset.
- Character encoding requiring the use of `encoding="latin1"` when importing the CSV files.

After these issues were addressed, the merged dataset was suitable for text preprocessing and machine learning.

```python
# Save final dataset
mental_health_df.to_csv("../data/processed/mental_health_text_dataset.csv", index=False)
```
